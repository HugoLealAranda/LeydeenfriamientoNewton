```python
import matplotlib.pyplot as plt
import numpy as np

# Leer el archivo CSV
df = pd.read_csv("C:\\Users\\hugo\\Documents\\lab6 termo\\aceite-agua.csv", delimiter=',', header=None, na_values='*')

# Extraer las columnas de interés
tiempo = df[1][29000:]  # Seleccionar a partir del segundo 0 al 250 en adelante
voltaje1 = df[2][29000:]  # Seleccionar a partir del segundo 0 al 250
voltaje2 = df[3][29000:]  # Seleccionar a partir del segundo 0 al 250

# Ajuste lineal para la curva de voltaje 1
coef_voltaje1 = np.polyfit(tiempo, voltaje1, 1)
pendiente_voltaje1 = coef_voltaje1[0]
intercepto_voltaje1 = coef_voltaje1[1]
ajuste_voltaje1 = np.polyval(coef_voltaje1, tiempo)
r2_voltaje1 = np.corrcoef(voltaje1, ajuste_voltaje1)[0, 1] ** 2

# Ajuste lineal para la curva de voltaje 2
coef_voltaje2 = np.polyfit(tiempo, voltaje2, 1)
pendiente_voltaje2 = coef_voltaje2[0]
intercepto_voltaje2 = coef_voltaje2[1]
ajuste_voltaje2 = np.polyval(coef_voltaje2, tiempo)
r2_voltaje2 = np.corrcoef(voltaje2, ajuste_voltaje2)[0, 1] ** 2

# Crear la gráfica
plt.plot(tiempo, voltaje1, label='Voltaje 1')
plt.plot(tiempo, ajuste_voltaje1, '--', label=f'Ajuste: y = {pendiente_voltaje1:.2f}x + {intercepto_voltaje1:.2f}\nR^2 = {r2_voltaje1:.2f}')

plt.plot(tiempo, voltaje2, label='Voltaje 2')
plt.plot(tiempo, ajuste_voltaje2, '--', label=f'Ajuste: y = {pendiente_voltaje2:.2f}x + {intercepto_voltaje2:.2f}\nR^2 = {r2_voltaje2:.2f}')

# Etiquetas de los ejes y título del gráfico
plt.xlabel('Tiempo')
plt.ylabel('Voltaje')
plt.title('Gráfico de voltajes en función del tiempo')

# Leyenda
plt.legend()

# Mostrar la gráfica
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos de calibración
Ti = 20.5  # Temperatura inicial (°C)
Tf = 60.5  # Temperatura final (°C)
Vi = 1.65  # Voltaje inicial (V)
Vf = 3.43  # Voltaje final (V)

# Puntos de calibración
temperaturas = [Ti, Tf]
voltajes = [Vi, Vf]

# Ajuste lineal
coeficientes = np.polyfit(temperaturas, voltajes, 1)
pendiente = coeficientes[0]
intercepto = coeficientes[1]

# Ecuación de la recta ajustada
ecuacion = f"V = {pendiente:.2f} * T + {intercepto:.2f}"

# Crear la gráfica
plt.scatter(temperaturas, voltajes, color='blue', label='Datos de calibración')
plt.plot(temperaturas, np.polyval(coeficientes, temperaturas), color='red', label='Ajuste lineal')
plt.xlabel('Temperatura (°C)')
plt.ylabel('Voltaje (V)')
plt.title('Calibración de instrumento: T1=V1')
plt.legend()

# Mostrar la ecuación en la gráfica
plt.text((Ti + Tf) / 2, (Vi + Vf) / 2, ecuacion, ha='center', va='bottom')

# Mostrar la gráfica
plt.show()
```

```python
from scipy.optimize import curve_fit
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Definición de la función de ajuste.
def func(t, T0, k):
    Ts = 25 # temperatura del ambiente
    return Ts + (T0 - Ts) * np.exp(-k * t)

# Leer el archivo CSV
df = pd.read_csv("C:\\Users\\hugo\\Documents\\lab6 termo\\aceite-agua.csv", delimiter=',', header=None, na_values='*')

# Extraer las columnas de interés
tiempo = df[1][:20000]
voltaje1 = df[2][:20000] 
voltaje2 = df[3][:20000] 

# Aplicar la ecuación de la calibración a la curva de voltaje 1
T_voltaje1 = 22.47 * voltaje1 - 16.58
# Aplicar la ecuación de la calibración a la curva de voltaje 2
T_voltaje2 = 22.47 * voltaje2 - 15.90

# Ajuste exponencial
popt1, pcov1 = curve_fit(func, tiempo, T_voltaje1)
popt2, pcov2 = curve_fit(func, tiempo, T_voltaje2)

# Graficación
plt.plot(tiempo, T_voltaje1, 'b-', label='T1=aceite')
plt.plot(tiempo, func(tiempo, *popt1), 'b--', label='fit: T0=%5.3f, k=%5.3f' % tuple(popt1))
plt.plot(tiempo, T_voltaje2, 'r-', label='T2=agua')
plt.plot(tiempo, func(tiempo, *popt2), 'r--', label='fit: T0=%5.3f, k=%5.3f' % tuple(popt2))
plt.xlabel('Tiempo [s]')
plt.ylabel('Temperatura [°C]')
plt.title('Gráfico de temperaturas en función del tiempo')
plt.legend()
plt.show()
```
